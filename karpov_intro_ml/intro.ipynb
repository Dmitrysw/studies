```python
#Как различается среднее и дисперсия оценок по предметам у групп студентов со стандартным или урезанным ланчем?
free_lunch_students = students_performance.loc[students_performance.lunch == "free/reduced" , ["math score", "writing score", "reading score"]]
standard_lunch_students = students_performance.loc[students_performance.lunch == "standard" , ["math score", "writing score", "reading score"]]
```

```python
#Средние оценки групп студентов с урезанным ланчем:
print("письмо студентов с урезанным ланчем - "+ str(free_lunch_students["writing score"].mean()))
print("математика студентов с урезанным ланчем - " + str(free_lunch_students["math score"].mean()))
print("чтение студентов с урезанным ланчем - " + str(free_lunch_students["reading score"].mean()))
```

```python
#Средние оценки групп студентов с полным ланчем:
print("письмо студентов с полным ланчем - "+ str(standard_lunch_students["writing score"].mean()))
print("математика студентов с полным ланчем - " + str(standard_lunch_students["math score"].mean()))
print("чтение студентов с полным ланчем - " + str(standard_lunch_students["reading score"].mean()))
```

```python
#Дисперсия групп студентов с урезанным ланчем:
print("дисперсия письмо студентов с урезанным ланчем - "+ str(free_lunch_students["writing score"].var()))
print("дисперсия математика студентов с урезанным ланчем - " + str(free_lunch_students["math score"].var()))
print("дисперсия чтение студентов с урезанным ланчем - " + str(free_lunch_students["reading score"].var()))
```

```python
#Дисперсия групп студентов с полным ланчем:
print("дисперсия письмо студентов с полным ланчем - "+ str(standard_lunch_students["writing score"].var()))
print("дисперсия математика студентов с полным ланчем - " + str(standard_lunch_students["math score"].var()))
print("дисперсия чтение студентов с полным ланчем - " + str(standard_lunch_students["reading score"].var()))
```
